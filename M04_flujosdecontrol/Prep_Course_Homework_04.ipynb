9) Imprimir los números primos existentes entre 0 y 30

```python

num = 2
while num < 30:
    for i in range(2,num+1):
        if num % i == 0:       #si se cumple que el resto entre el numero que nos dan y otro numero entre 2 y el numero anterior al numero dado es 0 entonces se para el primer bucle
            break
    if i == num:               #si el numero con el que se cumplio la condicion anterior es el mismo numero,entonces es primo                     
        print(f"{num} es primo")
    num +=1
```

10) ¿Se puede mejorar el proceso del punto 9? Utilizar las sentencias break y/ó continue para tal fin

Ciclo sin optimizar para comparar:

```python
tope = 4000
n = 2
primo= True
while n <= tope:
    for div in range(2, n):
        resto = n % div
        if resto == 0:
            primo = False
    if primo:
        print(n)
    else:
        primo = True
    n +=1
```

13) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
n = 99
while n <= 300:
    n += 1
    if n % 12 !=0: #si el numero n no es div por 12,entonces...
        continue    #...pasa al siguiente numero

    print(f"{n} es divisible por 12")
        
```
